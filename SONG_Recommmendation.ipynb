# artist file

```python
artists=pd.read_csv("spotify_artist_data_2023.csv")
artists=artists[['id','name']]
```

# Features file

```python
features=pd.read_csv("spotify_features_data_2023.csv").drop(columns=["key","instrumentalness","type","duration_ms","time_signature","analysis_url"],axis=1)
features.head(1)
# id danceability energy loudness mode speechiness acousticness valence  tempo track_href uri 
```

```python
tracks=pd.read_csv("spotify_tracks_data_2023.csv").drop("track_popularity",axis=1)
tracks.head(1)
# all
```

```python
tracks_and_features=tracks.merge(features,on="id")
tracks_and_features.head(1)
```

# Final Dataset

```python
songs=tracks_and_features.merge(albums,on="id")
songs=songs.merge(artists,on="artist_id").drop("label",axis=1)
songs.head(1)
```

# MinMax Scaler used for scale down each column

```python
from sklearn.preprocessing import MinMaxScaler
datatypes=['bool','float64', 'int64','object']
normalization=music_songs.select_dtypes(include=datatypes)
for col in normalization.columns:
    MinMaxScaler(col)
```

```python
normalization.describe()
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=10)
features=kmeans.fit_predict(normalization)
music_songs['features']=features
MinMaxScaler(music_songs['features'])
```
